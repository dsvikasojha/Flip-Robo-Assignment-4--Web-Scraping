Q1) Scrape the details of most viewed videos on YouTube from Wikipedia:
    Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
     You need to find following details:
        
        A) Rank
        
        B) Name
        
        C) Artist
        
        D) Upload date
        
        E) Views

```python
driver = webdriver.Chrome("C://chromedriver.exe") # Loading chrome driver
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos") # website from which we will scrape data
```

```python
details=[]

for i in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td"):
    details.append(i.text)
```

```python
rank=[]
name=[]
artist=[]
views=[]
upload=[]
count=0
#seperating details using indexing
while count<int(len(details)):
    rank.append(details[count])
    name.append(details[count+1])
    artist.append(details[count+2])
    views.append(details[count+3])
    upload.append(details[count+4])
    count=count+6 
```

```python
#creating dataframe
df=pd.DataFrame({})
df["Rank"]=rank
df["Name"]=name
df["Artist"]=artist
df['Upload date']=upload
df['Views']=views
df
```

Q 2) Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.

You need to find following details:

A) Match title (I.e. 1st ODI)

B) Series

C) Place

D) Date

E) Time Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver = webdriver.Chrome("C://chromedriver.exe") # Loading chrome driver
driver.get("https://www.bcci.tv/") # website from which we will scrape data
```

```python
# selecting fixtures option from international tab 
fixtures=driver.find_element_by_xpath("//div[@class='drop-down__options']/ul/li[1]/a").get_attribute("href")
driver.get(fixtures)
```

```python
#create empty list
match_title=[]
place=[]
series=[]
date=[]
time=[]
# scraping match title
for title in driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/p"):
    match_title.append(title.text.split("\n")[0])
# scraping place    
for p in driver.find_elements_by_xpath("//div[@class='fixture__description u-unskewed-text']/p"):
    place.append(p.text.split("\n")[1])
# scraping series
for s in driver.find_elements_by_xpath("//div[@class='fixture__format-strip']/span[2]"):
    series.append(s.text)        
# scraping date
for d in driver.find_elements_by_xpath("//div[@class='fixture__full-date']"):
    date.append(d.text.split("\n")[0:2])
#scraping time
for t in driver.find_elements_by_xpath("//div[@class='fixture__full-date']"):
    time.append(t.text.split("\n")[-1])
```

```python
series=pd.DataFrame({})
series["Match Title"]=match_title
series["Series"]=series
series["Place"]=place
series["Date"]=date
series["Time"]=time
series
```

Q4) Scrape the details of State-wise GDP of India from statisticstime.com.

Url = http://statisticstimes.com/

You have to find following details:

    A) Rank
    B) State
    C) GSDP(18-19)
    D) GSDP(17-18)
    E) Share(2017)
    F) GDP($ billion)

    Note: - From statisticstimes home page you have to reach to economy page through code.

```python
driver = webdriver.Chrome("C://chromedriver.exe") # Loading chrome driver
driver.get("http://statisticstimes.com/") # website from which we will scrape data
```

```python
# India's GDP
india=driver.find_element_by_xpath("//div[@class='navbar']/div[2]/div/a[3]").get_attribute("href")
driver.get(india)
```

```python
GDP = driver.find_element_by_xpath("//ul[@style = 'list-style-type:none;margin-left:20px;']/li/a").get_attribute('href')
driver.get(GDP)
```

```python
# Scrapping Details
ranks=[]
states=[]
GSDP_18_19=[]
GSDP_19_20=[]
share_18_19=[]
GDP_in_billion=[]

# Rank
for rank in driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[1]"):
    ranks.append(rank.text)
# States
for state in driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[2]"):
    states.append(state.text)
# GSDP for 19-20
for gsdp in driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[3]"):
    GSDP_19_20.append(gsdp.text)
# GSDP 18-19
for gsdp in driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[4]"):
    GSDP_18_19.append(gsdp.text)      
# Share    
for share in driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[5]"):
    share_18_19.append(share.text)    
# GDP in $ billion
for gdp in driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td[6]"):
    GDP_in_billion.append(gdp.text)
```

```python
df=pd.DataFrame({})
df["Rank"]=ranks
df["State"]=states
df["GSDP(19-20)"]=GSDP_19_20
df["GSDP(18-19)"]=GSDP_18_19
df["Share(18-19)"]=share_18_19
df["GDP($ billion)"]=GDP_in_billion
df
```

Q5) Scrape the details of trending repositories on Github.com.
    Url = https://github.com/
You have to find the following details:

    A) Repository title
    B) Repository description
    C) Contributors count
    D) Language used

```python
driver = webdriver.Chrome("C://chromedriver.exe") # Loading chrome driver
driver.get("https://github.com/ ") # website from which we will scrape data

# Selecting Trending option from from explore
trending=driver.find_element_by_xpath("//details[@class='HeaderMenu-details details-overlay details-reset width-full']/div/ul[2]/li[3]/a").get_attribute("href") 
driver.get(trending)
```

```python
#creating empty list to store scrapped data
rep_title=[]
rep_description=[]
cont_count=[]
lang_used=[]
url_list=[]
```

```python
# Respository title
for title in driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a"):
    rep_title.append(title.text)
```

```python
for url in driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a"):
    url_list.append(url.get_attribute("href"))
```

```python
# opening each repository
for i in url_list:
    driver.get(i)
    try:  # Repository Description
        description=driver.find_element_by_xpath("//div[@class='BorderGrid-cell']/p")
        rep_description.append(description.text)
    except NoSuchElementException as e:
        rep_description.append("-")
     
    try: # Contributors Count  
        count=driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[5]/div/h2/a[contains(text(),'Contributors ')]")
        cont_count.append(count.text.split(" ")[1])
    except NoSuchElementException as e:
        cont_count.append("-")
    
    try: # Language used
        lang=driver.find_element_by_xpath("//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[6]/div/ul")
        lang_used.append(lang.text.replace("\n"," "))
    except NoSuchElementException as e:
        lang_used.append("-")
```

```python
#create dataframe
df=pd.DataFrame({})
df["Repository title"]=rep_title
df["Repository description"]=rep_description
df["Contributors count"]=cont_count
df["Language used"]=lang_used
df
```

Q 9) Scrape the details most watched tv series of all time from imdb.com.

You have to find the following details:

A) Name

B) Year span

C) Genre

D) Run time

E) Ratings

F) Votes

```python
driver = webdriver.Chrome("C://chromedriver.exe") # Loading chrome driver
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
name = []
year = []
genre = []
run_time = []
ratings = []
votes = []
# Name
for i in driver.find_elements_by_xpath("//div[@class='lister-item-content']/h3/a"):
    name.append(i.text)
# Year
for j in driver.find_elements_by_xpath("//div[@class='lister-item-content']/h3/span[2]"):
    year.append(j.text)
# Genre
for k in driver.find_elements_by_xpath("//div[@class='lister-item-content']/p/span[5]"):
    genre.append(k.text)
# Run Time
for l in driver.find_elements_by_xpath("//div[@class='lister-item-content']/p/span[3]"):
    run_time.append(l.text)
# Votes
for n in driver.find_elements_by_xpath("//div[@class='lister-item-content']/div[1]/div/span[2]"):
    ratings.append(n.text)
# Ratings
for m in driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]"):
    votes.append(m.text)
```

```python
# Creating Dataframe
df = pd.DataFrame({})
df["Name"] = name
df["Year span"] = year
df["Genre"] = genre
df["Run time"] = run_time
df["Ratings"] = ratings
df["Votes"] = votes
df
```
